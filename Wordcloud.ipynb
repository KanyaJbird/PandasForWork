## สร้าง Word Cloud

```python
#เปลี่ยนรูปร่างของ mask

mask = np.array(Image.open(r"D:\ThaiPBS\รวมโค้ด\flag.png"))
```

```python
wordcloud = WordCloud(
                      font_path=path, 
                      relative_scaling = 0.3,
                      min_font_size=1,
                      background_color="white",
                      width=1024,
                      height=768,
                      max_words=200, 
                      mask=mask,
                      colormap='plasma', 
                      scale=3,
                      font_step=4,
                    #   contour_width=3, 
                    #   contour_color='steelblue', 
                      collocations=False,
                      regexp=regexp,
                      margin=2
                      ).generate(Kok)
```

แสดง Word Cloud

```python
fig, ax = plt.subplots(1, 1, figsize=(16, 12))
ax.imshow(wordcloud, interpolation='bilinear')
ax.axis("off")
fig.show()
```
